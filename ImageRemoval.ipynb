#ImageRemoval.ipynb

This code removes duplicate images of out datasets, which can occur as some images appear for more than one hashtag.

```python
import os
import csv
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
csv_file_path = '/content/drive/MyDrive/Klimawandel Projekt/Daten/INSTA_DATA_COMPLETE/klimawandel.csv'
image_folder_path = '/content/drive/MyDrive/Klimawandel Projekt/Daten/Updated_Insta_Daten/KLIMAWANDEL_CTDaten/klimawandel_images_names'
```

```python
# Read CSV file and extract ids
ids_in_csv = set()
with open(csv_file_path, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        id_value = row['Id'].replace('.jpg', '')  # Remove ".jpg" from the ID
        ids_in_csv.add(id_value)
        #ids_in_csv.add(row['Id'])
```

```python
counter = 0
# Iterate through images in the folder
for image_filename in os.listdir(image_folder_path):
    image_path = os.path.join(image_folder_path, image_filename)

    # Extract id from image filename
    image_id, _ = os.path.splitext(image_filename)
    print(image_id)

    # Check if the image id is not in the CSV
    if image_id not in ids_in_csv:
        # Remove the image file
        os.remove(image_path)
        counter = counter +1
        print(f"Removed image: {image_path}")

print("Image removal process completed.")
print(counter)
```
