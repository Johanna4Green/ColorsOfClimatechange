This Notebook compares 3 different annotations of the same images (represented in the table by the column 'annotation_id') for our Klimawandel-Project. The choices columns reflect the different Annotation categories:



*   Correlation to the Topic: choice1
*   Type of Media: choice2
*   Impression/Sentiment: choice3
*   Animals: choice4
*   Nature: choice5
*   People: choice6
*   Destruction of Nature/ Natural Disaster: choice7
*   Artifact: choice8
*   Education/Facts/News: choice9

Every time there was at least one inconsistency between the 3 annotators, an x appears in the table.

```python
import pandas as pd
from tabulate import tabulate

file1 = '/content/drive/MyDrive/Klimawandel Projekt/Annotation/LabelStudio_Annotationen/Klimawandel1 CSV.csv'
file2 = '/content/drive/MyDrive/Klimawandel Projekt/Annotation/LabelStudio_Annotationen/Klimawandel2 CSV.csv'
file3 = '/content/drive/MyDrive/Klimawandel Projekt/Annotation/LabelStudio_Annotationen/Klimawandel3 CSV.csv'

def compare_csv(file1, file2, file3):
    # CSV-Dateien einlesen
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    df3 = pd.read_csv(file3)

    # Die ersten 30 Zeilen auswhlen
    df1_sample = df1.head(1200)
    df2_sample = df2.head(1200)
    df3_sample = df3.head(1200)

    # Spalten fr den Vergleich auswhlen (hier gehe ich von den Spalten 3 bis Ende aus, bitte anpassen)
    compare_cols = df1_sample.columns[2:]

    # Vergleiche jede Zeile einzeln fr alle Annotatoren, beginnend ab der dritten Spalte
    diff_1_2 = df1_sample[compare_cols].ne(df2_sample[compare_cols])
    diff_1_3 = df1_sample[compare_cols].ne(df3_sample[compare_cols])
    diff_2_3 = df2_sample[compare_cols].ne(df3_sample[compare_cols])

    # Kombiniere die Unterschiede in einem DataFrame
    all_diffs = diff_1_2 | diff_1_3 | diff_2_3

    # Fge die ersten beiden Spalten zu den Unterschieden hinzu
    all_diffs = pd.concat([df1_sample.iloc[:, :2], all_diffs], axis=1)

    # Erstelle eine Tabelle mit Unterschieden markiert
    table = tabulate(all_diffs.replace({True: 'X', False: ''}), headers='keys', tablefmt='pretty', showindex=False)

    # Zeige die Tabelle mit Unterschieden
    print(f"Unterschiede zwischen {file1}, {file2} und {file3}:\n")
    print(table)

# Beispielaufruf mit den Dateinamen deiner CSV-Dateien
compare_csv(file1, file2, file3)


```
