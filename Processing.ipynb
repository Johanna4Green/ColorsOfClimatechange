```python
total_total_df = pd.merge(total_df, color_df_col, on='id', how='outer')
```

```python
import re

def extract_numbers(text):
    numbers = re.findall(r'\d+', text)  # Extract numbers using regular expression
    return ''.join(numbers)  # Join the extracted numbers into a single string

def update_platform(row):
    if pd.isna(row['platform']):  # Check if platform is NaN
        id_numbers = extract_numbers(row['id'])  # Extract numbers from id column
        if len(id_numbers) >= 5:  # Check if id has at least 5 numbers
            return 'tiktok'
        else:
            return 'instagram'
    else:
        return row['platform']  # If platform is not NaN, return its original value

# Apply the function to update the 'platform' column
total_total_df['platform'] = total_total_df.apply(update_platform, axis=1)
```

# JSON

```python
import pandas as pd
import json

# Load the JSON data
with open('/content/drive/MyDrive/Uni/SocialMedSeminar/Real_Daten/Klimawandel3 JSON.json', 'r') as file:
    data = json.load(file)

# Initialize an empty list to store processed dictionaries
processed_data = []

# Iterate over each dictionary in the list
for item in data:
    # Extract the last part of the "image" value and remove ".jpg" and ".jpeg"
    image_value = item['data']['image'].split('/')[-1].replace('.jpg', '').replace('.jpeg', '')

    # Add a new key-value pair for the "ID" column
    item['ID'] = image_value

    # Append the modified dictionary to the processed_data list
    processed_data.append(item)

# Save the modified list of dictionaries back to a JSON file
with open('/content/drive/MyDrive/Uni/SocialMedSeminar/Real_Daten/modified_json_file3.json', 'w') as file:
    json.dump(processed_data, file, indent=4)

# Create a DataFrame from the list of processed dictionaries (optional)
df = pd.DataFrame(processed_data)

```
