# Instagram Pictures Download

The following plan:


1.   Go through the column "Photo" of the csv-file which contains the link to the Instagram Post Picture
2.   With Request open Picture in the Browser
3.   Download the Picture and store it in a folder. The image should be numbered serially to later connect them to the row in the the csv-file.

```python
# Imports libraries
import os
import pandas as pd
import requests
from google.colab import drive
from google.colab import files
```

```python
# Connect to Google Drive
drive.mount('/content/drive')
```

```python
# Set up the folder structure
base_folder = '/content/drive/My Drive/CLIMATECRISIS_CTDaten'
pics_folder = 'climatecrisis_images'
pics_path = os.path.join(base_folder, pics_folder)

# Create the folder if it doesn't exist
if not os.path.exists(pics_path):
    os.makedirs(pics_path)
```

```python
# Load the CSV dataset
csv_path = os.path.join(base_folder, 'climatecrisis.csv')
print(csv_path)
df = pd.read_csv("climatecrisis.csv", sep=';')
```

```python
# check if the Photo Column really contains the link to the picture
df["Photo"][2]
```

```python
# Iterate through the Photo column and download images
for index, row in df.iterrows():
    photo_url = row['Photo']

    try:
        # Use requests library to download images
        response = requests.get(photo_url)
        response.raise_for_status()  # Raise an error for bad responses

        # Save the image with a sequential filename
        new_filename = f"{index + 1}.jpg"
        new_filepath = os.path.join(pics_path, new_filename)

        with open(new_filepath, 'wb') as f:
            f.write(response.content)

        # add 2 so that the picture-number/ picture-name is the same as the row in the csv file
        # for the english pictures this is only + 1 (see documentation of english Instagram Pictures)
        print(f"Downloaded image for row {index + 2}")

    except requests.exceptions.RequestException as e:
        # get the images that failed to be able to double check with the url in the URL Column
        print(f"IMAGE DOWNLOAD FAILED FOR ROW{index + 2}. ERROR: {e}")
```
